```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(0)
n = 100
x = np.random.normal(0, 1, n)
y = np.random.normal(0, 1, n)
categories = np.random.choice(['A', 'B', 'C'], n)

# Plot the scatter plot
plt.scatter(x, y, c=categories, cmap='viridis')
plt.colorbar()
plt.show()
```

```python
import plotly.offline as pyo
import plotly.graph_objects as go

tips = px.data.tips(pretty_names=True)

trace = go.Scatter(
    x=tips['Total Bill'],
    y=tips['Tip'],
    mode='markers',
    marker=dict(
        size=5,
        color='rgb(10,112,32)',
        symbol='pentagon'
    )
)

layout = go.Layout(
    title='Tip by Total Bill',
    xaxis=dict(
        title='Total Bill'
    ),
    yaxis=dict(
        title='Tip'
    )
)

data = [trace]
fig = go.Figure(data=data, layout=layout)
pyo.plot(fig, filename='temp-plot.html')
```
